```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about Artificial Intelligence and its applications",
    candidate_labels=["AI", "machine learning", "robotics", "data science"],
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("Throughout this journey, we’ll guide you to")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "This journey will transform you into",
    max_length=30,
    num_return_sequences=2,
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will transform you into an <mask> master", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Angel and I work at Infinite Learning in Batam City.")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Machine Learning adalah cabang dari kecerdasan buatan yang belajar dari data tanpa diprogram secara eksplisit."
question = "Apa itu Machine Learning?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("Artificial Intelligence is revolutionizing industries and reshaping the future.")
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    Indonesia has undergone significant changes in recent decades. The country has seen a rapid increase in urbanization, with a growing number of cities and a shift towards a more modern lifestyle. However, there are still challenges that remain, such as poverty, inequality, and environmental issues. The Indonesian government has implemented various policies to address these issues, including initiatives to improve education, healthcare, and infrastructure.

    In recent years, Indonesia has also seen a boom in digital technology and e-commerce, with many startups emerging in the tech industry. The growth of this sector has created new job opportunities, but it has also highlighted the need for more skilled workers in technology and engineering fields. The country continues to face challenges in terms of access to quality education, especially in rural areas.

    While Indonesia is making progress in various sectors, it still faces obstacles related to governance, corruption, and the unequal distribution of resources. Efforts to improve the country's overall development are ongoing, with a focus on creating sustainable economic growth and addressing social issues.
    """
)
```

```python
from transformers import pipeline

translator = pipeline("translation_id_to_en", model="Helsinki-NLP/opus-mt-id-en")

text_to_translate = "Saya sangat membenci durian karena baunya sangat tidak enak"
result = translator(text_to_translate)

print(result[0]['translation_text'])
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering")
context = "Universitas Negeri Padang (UNP) adalah sebuah perguruan tinggi negeri yang terletak di kota Padang."
question = "Apa itu Universitas Negeri Padang?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

# Analisis

Kode yang diberikan menggunakan pustaka transformers dari Hugging Face untuk mengimplementasikan berbagai aplikasi pemrosesan bahasa alami (NLP) menggunakan model pre-trained. Setiap bagian kode mengaktifkan pipeline tertentu yang memungkinkan untuk menyelesaikan berbagai tugas tanpa memerlukan pelatihan tambahan.

- Pertama, kode menggunakan zero-shot classification untuk mengklasifikasikan teks ke dalam beberapa kategori tanpa pelatihan model khusus. Ini dilakukan dengan memberikan kandidat label yang relevan dan membiarkan model memilih yang paling sesuai.

- Selanjutnya, pipeline text-generation digunakan untuk menghasilkan kelanjutan teks berdasarkan input awal, dan dengan menggunakan model yang lebih kecil (DistilGPT-2), kode ini menghasilkan teks dengan panjang dan jumlah urutan yang dikontrol.

- Kode juga mencakup fill-mask, yang memungkinkan model untuk mengisi kata yang hilang dalam teks, seperti menggantikan <mask> dengan kata yang paling cocok.

- Untuk pengenalan entitas, pipeline NER mengidentifikasi nama orang, tempat, dan organisasi dalam teks, membantu ekstraksi informasi penting dari kalimat.

- Selain itu, kode memanfaatkan pipeline question-answering untuk menjawab pertanyaan terkait dengan konteks yang diberikan, yang memungkinkan model untuk memberikan jawaban berdasarkan pemahaman dari teks yang disediakan.

- Untuk menganalisis sentimen dari teks, pipeline sentiment-analysis digunakan, yang menentukan apakah teks memiliki sentimen positif atau negatif.

- Kode ini juga mencakup pipeline summarization yang merangkum teks panjang menjadi bentuk yang lebih ringkas dan padat.

- Terakhir, pipeline translation digunakan untuk menerjemahkan teks dari Bahasa Indonesia ke Bahasa Inggris, memberikan contoh aplikasi dalam domain penerjemahan antarbahasa.

Secara keseluruhan, kode ini mencakup berbagai fungsi NLP yang memungkinkan untuk menganalisis, menghasilkan, dan memanipulasi teks dengan cara yang sangat efisien menggunakan model yang sudah dilatih sebelumnya. Ini menunjukkan fleksibilitas dan kemudahan penggunaan pustaka transformers untuk aplikasi-aplikasi praktis dalam analisis bahasa.
